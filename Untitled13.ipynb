```python
df =pd.read_csv('practice-csv.csv')
df.fillna({'Pulse':20},inplace=True)
print(df.to_string())
```

```python
x=df['Calories'].mean()
df.fillna({'Calories':x}, inplace=True)

print(df.to_string()  )
```

```python
x=df["Cost"].mode()[0]
print(x)
df.fillna({'Cost':x},inplace=True)
print(df.to_string())
```

```python
x=df['Maxpulse'].median()
print(x)
df.fillna({'Maxpulse':x},inplace=True)
print(df.to_string())
```

```python
df['Date']= pd.to_datetime(df['Date'], format='mixed')
print(df.to_string())
```
